```
ads_data = pd.read_csv('/dbfs/FileStore/tables/Social_Network_Ads.csv')
ads_data.head()
```

```
#Check the info of the Data (Null cells if any)
ads_data.info()
```

```
# Get dummies for the categorical variables
Male = pd.get_dummies(ads_data['Gender'],drop_first=True)
ads_data = pd.concat([ads_data,Male],axis=1)
ads_data
```

```
# Split the Dependable Variables(That we want to predict) and Independent Variable(Our Features)
X= ads_data.drop(['Gender','Purchased'],axis=1)
Y= ads_data['Purchased']
```

```
# Split the train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=50)
```

```
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```
# Create a Logistic Regression Model and fit the train data
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train,Y_train)
```

```
# predict the test set
pred_log = log_model.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred_log))
```

```
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(X_train,Y_train)
```

```
knn_pred = KNN.predict(X_test)
```

```
print(classification_report(Y_test,knn_pred))
```
